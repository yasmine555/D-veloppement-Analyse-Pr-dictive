```python
import os
import pandas as pd
import plotly.express as px
from dotenv import load_dotenv

load_dotenv()
API_KEY = os.getenv("OPENWEATHER_API_KEY")

```

```python
import sys
sys.path.append(os.path.abspath("../src"))

from weather_connector import get_weather_forecast

```

```python
# 4. Tester avec une ville
ville = "Tunis"
forecast_data = get_weather_forecast(ville, API_KEY)

df = pd.DataFrame(forecast_data)
df['datetime'] = pd.to_datetime(df['datetime'])

df.head()

# 7. Visualisation température dans le temps
fig = px.line(df, x="datetime", y="temp", title=f"Prévisions météo à {ville}", markers=True)
fig.update_layout(yaxis_title="Température (°C)", xaxis_title="Date/Heure")
import plotly.io as pio
pio.renderers.default = "browser"
fig.show()

#  Affichage des descriptions météo
df[['datetime', 'description']].head(10)
```
