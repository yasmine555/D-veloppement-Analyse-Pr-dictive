```python
# ðŸ”§ Correction pour que Jupyter trouve le dossier 'src'
import sys
import os
sys.path.append(os.path.abspath("../src"))

# ðŸ“¦ Imports
from telco_loader import load_telco_dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ§¹ Charger les donnÃ©es
df = load_telco_dataset("../data/WA_Fn_UseC_Telco_Customer_Churn.csv")

# ðŸ‘€ AperÃ§u
print("Dimensions :", df.shape)
df.head()
```

```python
#RÃ©partition du churn (Yes/No)
sns.countplot(x="Churn", data=df)
plt.title("RÃ©partition des clients : Churn vs Non-Churn")
plt.xlabel("Churn")
plt.ylabel("Nombre de clients")
plt.show()

#Churn selon le contrat

sns.countplot(x="Churn", hue="Contract", data=df)
plt.title("Churn selon le type de contrat")
plt.xlabel("Churn")
plt.ylabel("Nombre de clients")
plt.xticks([0, 1], ["Non", "Oui"])
plt.show()
```

```python
#3. Heatmap de corrÃ©lations numÃ©riques

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("CorrÃ©lations numÃ©riques")
plt.show()
```

```python
#4. Ã‚ge du client (tenure) et churn

sns.histplot(data=df, x="tenure", hue="Churn", bins=30, kde=True)
plt.title("Distribution de l'anciennetÃ© client (tenure) selon churn")
plt.show()
```
